```python
import schemdraw as schem
import schemdraw.elements as e
import matplotlib.pyplot as plt
import math



%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
d = schem.Drawing()

U = d.add(e.Battery, d = 'up', label = 'U')
R1 = d.add(e.RES, d = 'right', label = '$R_1$')
R2 = d.add(e.RES, label = '$R_2$')
C1 = d.add(e.CAP, d = 'down', xy = R1.end, label = '$C_1$' )
d.add(e.LINE, d = 'right', xy = U.start, l = 2)
C2 = d.add(e.CAP, label = '$C_1$', l = 1 )


d.draw()
```

```python
d = schem.Drawing(unit = 5)
U = d.add(e.SourceV, d = 'up',  lblofst=0.3)

R1 = d.add(e.RES, d = 'right', label = '$R_1$')
R2 = d.add(e.RES, d = 'down', label = '$R_2$')
d.add(e.LINE, d = 'right', xy = R2.start, l = 2)

R3 = d.add(e.RES, d = 'down', label = '$R_3$')
d.add(e.LINE, d = 'right', xy = R3.start, l = 2)

R4 = d.add(e.RES, d = 'down', label = '$R_4$', l = 2.5)
R5 = d.add(e.RES, d = 'down', botlabel = '$R_5$', l = 2.5)

d.add(e.LINE, d = 'left', xy = R5.end, l = 2)
d.add(e.LINE, d = 'left', xy = R3.end, l = 2)

R6 = d.add(e.RES, d = 'left', label = '$R_6$', l = 2.5)
R7 = d.add(e.RES, d = 'left', label = '$R_7$', l = 2.5)
R8 = d.add(e.RES, theta = 45, label = '$R_8$', to = R1.end )

d.draw()
```
